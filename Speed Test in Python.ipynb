```python
import subprocess
```

```python
def internet_speed_test():
    # Run the speedtest-cli command
    command_output = subprocess.run(['speedtest-cli', '--simple'], capture_output=True, text=True)

    # Retrieve the download and upload speeds from the command output
    output_lines = command_output.stdout.strip().split('\n')
    download_speed = float(output_lines[1].split()[1])
    upload_speed = float(output_lines[2].split()[1])

    # Print the results
    print(f"Download Speed: {download_speed:.2f} Mbps")
    print(f"Upload Speed: {upload_speed:.2f} Mbps")
```

```python
# Call the internet_speed_test function
internet_speed_test()
```
